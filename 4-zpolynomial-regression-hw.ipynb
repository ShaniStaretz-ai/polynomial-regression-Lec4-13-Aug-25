```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
# Our data
x_fertilizer_Kg = np.array([50, 100, 150, 200, 250, 300, 350, 400, 450, 500]).reshape(-1, 1)
y_Potatoes_t = np.array([7.5, 10.2, 12.8, 14.5, 15.6, 16.0, 15.8, 15.0, 13.5, 11.2])
```

```python
poly_model = Pipeline([
    ('poly_features', PolynomialFeatures(degree=2)),
    ('linear_regression', LinearRegression())
])
poly_model.fit(x_fertilizer_Kg, y_Potatoes_t)
y_poly_pred = poly_model.predict(x_fertilizer_Kg)
```

```python
# Plotting
x_range = np.linspace(50, 500, 200).reshape(-1, 1)
plt.figure(figsize=(10,6))
plt.scatter(x_fertilizer_Kg, y_Potatoes_t, color='black', label='Data points')

# Polynomial curve
plt.plot(x_range, poly_model.predict(x_range), color='red', label='Polynomial Regression (deg 2)')

plt.xlabel("Fertilizer (Kg)")
plt.ylabel("Potato Yield (t)")
plt.title("Potato Yield vs Fertilizer")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Linear fit
x_fertilizer_Kg = np.array([50, 100, 150, 200, 250, 300, 350, 400, 450, 500])

# Polynomial fit (degree 2)
poly_coeffs = np.polyfit(x_fertilizer_Kg, y_Potatoes_t, deg=2)
poly_fit = np.poly1d(poly_coeffs)
print(f"Polynomial equation: y = {poly_coeffs[0]:.6f}x + {poly_coeffs[1]:.4f}x + {poly_coeffs[2]:.4f}")

# -----------------------------
# Plotting
x_range = np.linspace(50, 500, 200)
plt.figure(figsize=(10,6))
plt.scatter(x_fertilizer_Kg, y_Potatoes_t, color='black', label='Data points')
# Polynomial
plt.plot(x_range, poly_fit(x_range), color='red', label='Polynomial fit (deg 2)')

plt.xlabel("Fertilizer (Kg)")
plt.ylabel("Potato Yield (t)")
plt.title("Potato Yield vs Fertilizer")
plt.legend()
plt.grid(True)
plt.show()
```

```python
poly_coeffs = np.polyfit(x_fertilizer_Kg, y_Potatoes_t, deg=2)
a, b, c = poly_coeffs
x_optimal = -b / (2*a)
y_max = poly_fit(x_optimal)
print(f"Optimal fertilizer: {x_optimal:.1f} Kg -> Maximum yield: {y_max:.2f} t")
```
